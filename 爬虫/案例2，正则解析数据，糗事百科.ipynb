```python
import requests
from urllib.request import urlretrieve
import re
import os

headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.72 Safari/537.36"
}
dir_name = 'img'
if not os.path.exists(dir_name):
    os.mkdir(dir_name)

for i in range(1, 4):
    url = f"https://www.qiushibaike.com/imgrank/page/{i}/"
    response = requests.get(url, headers=headers)
    try:
        if response.status_code == 200:
            page_html = response.text
        else:
            page_html = None
            response.raise_for_status()
    except :
        print('出现异常')

    if page_html:
        pattern = '<div class="thumb">.*?<img src="(.*?)" alt=.*?</div>'
        img_src_list = re.findall(pattern, page_html, re.S)
        img_url_list = ['https:' + item for item in img_src_list]
        for item in img_url_list:
            img_name = item.split('/')[-1]
            img_path = dir_name + '/' + img_name
            urlretrieve(item, img_path)
            print(img_name, '下载完成')
```
